```python
# 2. Encode data
# since there are some unknown values in several of the categorical variables we need to decide how to treat them. 
# In class we discussed some options:
# 1. We could drop all the unknown values. This has the advantage of creating a dataset with the complete data.
    # The disadvantage is that we are exluding obeservations. Future datasets may include unknown values which will also need to be dropped.
# 2. We could treat the unknown values as the omitted variable. 
    # This will preserve dataset length and give us a way to handle uknown values in the future. 
    # But setting our omitted value to unknown values could confound the results. 
# 3. For missing or unknown numeric values we can impute the values using the mean or some other value the would be appropriate.
    # Since I dont see any missing numeric data we can ignore this option.
    
# Since our objective is to predict whether or not a product is purchased based on the available data of a prospective client then 
# we should drop all the unknown values in the data. 
# If after the cleaning process we lose significant observations we can reconsider this approach.

bank.head()
```

```python
# Create a dictionary with the categorical data points for each row
# it looks like the OneHotEncoder does the legwork to create dummy variables for all the categorical variables
# start with the kitchen sink approach and convert all the data.

cat_columns = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome', 'y']
cat_dict = bank[cat_columns].to_dict(orient='records')
#cat_dict[:5]
```

```python
# Construct a DictVectorizer to transform our dictionary to
# a binary one-hot encoded array for each row
vec = feature_extraction.DictVectorizer()
cat_vector = vec.fit_transform(cat_dict).toarray()
#cat_vector[:5]
```

```python
# Construct a separate dataframe with the one-hot encoded data
# and set the column names by calling get_feature_names
df_vector = pd.DataFrame(cat_vector)
vector_columns = vec.get_feature_names()
#vector_columns
```

```python
# Drop the categorical columns and join the new one-hot 
# dataframe with the original dataframe
df_vector.columns = vector_columns
df_vector.index = bank.index

bank = bank.drop(cat_columns, axis=1)
bank = bank.join(df_vector)
bank.head()
```

```python

#bank.describe()
bank.info()
```

```python
# drop the unknown data rows for all variables that are unknown

# create an object that captures the variables contain data that we need to drop.

feature_cols = ['contact=unknown', 'education=unknown', 'job=unknown', 'poutcome=unknown']

#bank_test = bank[bank['contact=unknown']!=1]
#bank_test.head()

bank_test = bank
for x in feature_cols:
    bank_test = bank_test[bank_test[x]!=1]



#bank_test.info()

# Dropping all of the unknown values has left us with 764 obs which are not very many so treat them as the omitted var.
```

```python
# create a list of feature columns
feature_cols = ['age', 'job=admin.', 'job=blue-collar', 'job=entrepreneur', 'job=housemaid', 'job=management', 'job=retired',
                'job=self-employed', 'job=services', 'job=student', 'job=technician', 'job=unemployed', 
                'marital=single', 'education=primary', 'education=secondary', 'education=tertiary', 'default=yes',  
                'housing=yes', 'loan=yes', 'contact=cellular', 'contact=telephone', 'pdays', 'previous', 'poutcome=failure', 
                'poutcome=other', 'poutcome=success', 'month=feb', 'month=mar', 'month=apr', 'month=may', 'month=jun', 'month=jul', 
                'month=aug', 'month=sep', 'month=oct', 'month=nov', 'month=dec']

# define X and y
features = bank[feature_cols].values
target = bank['y=yes'].values
# define X (features) and y (response)
X = bank[feature_cols]
y = bank['y=yes']

# split into train/test
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
treeclf = DecisionTreeClassifier(max_depth=3, random_state=1)
treeclf.fit(X_train, y_train)

preds = treeclf.predict(X_test)
np.sqrt(metrics.mean_squared_error(y_test, preds))
```

```python
pd.DataFrame({'feature':feature_cols, 'importance':treeclf.feature_importances_}).sort('importance', ascending = False)
```

```python
# fit a decision tree with all the data
treeclf = DecisionTreeClassifier(max_depth=3, random_state=1)

treeclf.fit(X, y)

preds = treeclf.predict(X)
rmseb = np.sqrt(metrics.mean_squared_error(y, preds))
rmseb
```

```python
# 4. Evaluate the accuracy of the decision tree. 
scores = cross_val_score(treeclf, X, y, cv = 5, scoring = 'accuracy')
#scores = cross_val_score(treeclf, X_train, y_train, cv = 5, scoring = 'accuracy')
cvs_b = np.mean(scores)
cvs_b
```
